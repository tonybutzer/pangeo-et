```python
def print_du(url_path):
    disk_usage = fs.du(url_path)
    hf_disk_usage = humanfriendly.format_size(disk_usage)
    print(url_path, " : ", hf_disk_usage)
```

```python
for psuedo_dir in subdir_list:
    print(psuedo_dir)
    print_du(psuedo_dir)
```
